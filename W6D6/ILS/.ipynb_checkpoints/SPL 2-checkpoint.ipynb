```python
# Create sample data:
data = [
    (1, "Alice", "Sales", 5000),
    (2, "Bob", "IT", 6000),
    (3, "Charlie", "Finance", 5500),
    (4, "Diana", "Sales", 6500),
    (5, "Evan", "IT", 7000)
]

columns = ["id", "name", "dept", "salary"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
# Inspect schema and structure:
df.printSchema()
df.describe().show()
```

```python
# Select specific columns:
df.select("name", "salary").show()
```

```python
# Rename a column:
df_renamed = df.withColumnRenamed("dept", "department")
df_renamed.show()
```

```python
# Filter rows where salary > 6000:
df.filter(df.salary > 6000).show()
```

```python
# Apply multiple conditions:
df.filter((df.salary > 5500) & (df.dept == "IT")).show()
```

```python
# Sort by salary (ascending and descending):
df.orderBy("salary").show()
df.orderBy(df.salary.desc()).show()
```

```python
# Import helper functions:
from pyspark.sql.functions import col, when, lit
```

```python
# Add a new derived column:
df_updated = df.withColumn("bonus", df.salary * 0.1)
df_updated.show()
```

```python
# Create conditional columns:
df_cond = df_updated.withColumn(
    "performance",
    when(col("salary") > 6500, "High")
    .when(col("salary") > 5500, "Medium")
    .otherwise("Low")
)
df_cond.show()
```

```python
# Calculate average salary per department:
from pyspark.sql.functions import avg

df.groupBy("dept").agg(avg("salary").alias("avg_salary")).show()
```

```python
# Count employees per department:
df.groupBy("dept").count().show()
```

```python
# Create a second DataFrame for bonus details:
bonus_data = [
    (1, 500),
    (2, 800),
    (3, 700),
    (4, 900),
    (5, 1000)
]
bonus_df = spark.createDataFrame(bonus_data, ["id", "bonus"])
```

```python
# Perform an inner join:
joined_df = df.join(bonus_df, on="id", how="inner")
joined_df.show()
```

```python
# Union two DataFrames with identical schema:
union_df = df.union(df)
union_df.show()
```

```python
# Remove duplicates:
df_distinct = df.dropDuplicates(["dept"])
df_distinct.show()
```

```python
# Add a null record (for testing):
df_null = df.withColumn("dept", when(col("dept") == "Sales", None).otherwise(col("dept")))
df_null.show()
```

```python
# Drop null values:
df_clean = df_null.na.drop()
df_clean.show()
```

```python
# Fill missing values:
df_filled = df_null.na.fill({"dept": "Unknown"})
df_filled.show()
```

```python
# Trigger computation with an action:
df.count()
```

```python
# View the logical and physical execution plan:
df.explain(True)
```

```python
# Write the cleaned DataFrame to disk:
df_clean.write.mode("overwrite").csv("output/cleaned_data")
```

```python
# Read it back:
spark.read.csv("output/cleaned_data").show()
```
