## Phase 2 – Load Data into Spark DataFrames

```python
from pyspark.sql import SparkSession

# Create a SparkSession if one doesn't exist
spark = SparkSession.builder.appName("SQLOperationsLab").getOrCreate()

print("SparkSession created successfully!")
```

```python
custDF = spark.read.option("header", True).option("inferSchema", True).csv("customers.csv")
salesDF = spark.read.option("header", True).option("inferSchema", True).csv("sales.csv")

print("Customers DataFrame:")
custDF.show()

print("Sales DataFrame:")
salesDF.show()
```

## Phase 3 – Register as Temporary SQL Views

```python
custDF.createOrReplaceTempView("customers")
salesDF.createOrReplaceTempView("sales")

print("Temporary views 'customers' and 'sales' registered successfully!")
```

## Phase 4 – Run Basic SQL Queries

```python
print("Querying all data from customers:")
spark.sql("SELECT * FROM customers").show()

print("Querying all data from sales:")
spark.sql("SELECT * FROM sales").show()
```

```python
print("Selecting specific columns (name, region) from customers:")
spark.sql("SELECT name, region FROM customers").show()
```

## Phase 5 – Filter Data

```python
print("Retrieving transactions greater than $200:")
spark.sql("SELECT * FROM sales WHERE amount > 200").show()
```

```python
print("Filtering transactions by category 'Books':")
spark.sql("SELECT * FROM sales WHERE category='Books'").show()
```

## Phase 6 – Perform Sorting and Ordering

```python
print("Sorting transactions by amount in descending order:")
spark.sql("SELECT * FROM sales ORDER BY amount DESC").show()
```

```python
print("Getting the top 3 highest transactions:")
spark.sql("SELECT * FROM sales ORDER BY amount DESC LIMIT 3").show()
```

## Phase 7 – Aggregations and Grouping

```python
print("Finding total and average sales per category:")
spark.sql("""
SELECT category, SUM(amount) AS total_sales, AVG(amount) AS avg_sales
FROM sales
GROUP BY category
ORDER BY total_sales DESC
""").show()
```

## Phase 8 – Join Multiple Tables

```python
print("Combining customer and sales data using an inner join:")
spark.sql("""
SELECT c.name, c.region, s.category, s.amount
FROM customers c
JOIN sales s
ON c.cust_id = s.cust_id
""").show()
```

## Phase 9 – Complex SQL Operations

```python
print("Adding conditional logic with CASE WHEN to categorize purchases:")
spark.sql("""
SELECT c.name, s.amount,
CASE
  WHEN s.amount > 300 THEN 'Premium Purchase'
  WHEN s.amount BETWEEN 150 AND 300 THEN 'Standard Purchase'
  ELSE 'Basic Purchase'
END AS purchase_type
FROM customers c
JOIN sales s
ON c.cust_id = s.cust_id
""").show()
```

## Phase 10 – Create Aggregated Insights

```python
print("Finding total sales and transactions per region:")
spark.sql("""
SELECT c.region, SUM(s.amount) AS total_sales, COUNT(s.txn_id) AS transactions
FROM customers c
JOIN sales s
ON c.cust_id = s.cust_id
GROUP BY c.region
ORDER BY total_sales DESC
""").show()
```

## Bonus – Create a Temporary Analytical View

```python
print("Creating a temporary view 'customer_summary' with total sales and customer rank:")
spark.sql("""
CREATE OR REPLACE TEMPORARY VIEW customer_summary AS
SELECT
  c.cust_id,
  c.name,
  c.region,
  SUM(s.amount) AS total_sales,
  COUNT(s.txn_id) AS total_transactions,
  RANK() OVER (ORDER BY SUM(s.amount) DESC) as sales_rank
FROM customers c
JOIN sales s
  ON c.cust_id = s.cust_id
GROUP BY
  c.cust_id,
  c.name,
  c.region
ORDER BY
  total_sales DESC
""").show()

print("Displaying the 'customer_summary' view:")
spark.sql("SELECT * FROM customer_summary").show()
```
