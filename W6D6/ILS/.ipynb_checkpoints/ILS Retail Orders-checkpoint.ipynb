```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("RetailOrdersLab").getOrCreate()
```

```python
from pyspark.sql.types import (
    StructType, StructField,
    IntegerType, StringType, DoubleType
)

schema = StructType([
    StructField("order_id", IntegerType(), True),
    StructField("order_date", StringType(), True),
    StructField("customer_id", StringType(), True),
    StructField("product", StringType(), True),
    StructField("category", StringType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("unit_price", DoubleType(), True),
    StructField("payment_method", StringType(), True),
    StructField("store_city", StringType(), True)
])


#Read data from csv using defined schema
df = spark.read.option("header", "true").schema(schema).csv("orders.csv")
```

```python
df.printSchema()
df.show(5)
```

```python
from pyspark.sql.functions import col, to_date

df = df.withColumn("order_date", to_date(col("order_date")))
```

```python
df = df.withColumn("total_amount", col("quantity") * col("unit_price"))
```

```python
df = df.withColumnRenamed("store_city", "city").withColumnRenamed("unit_price", "price")
```

```python
from pyspark.sql.functions import sum

category_revenue = df.groupBy("category").agg(sum("total_amount").alias("category_revenue"))

category_revenue.show()
```

```python
from pyspark.sql.functions import count

city_stats = df.groupBy("city").agg(count("order_id").alias("total_orders"),sum("total_amount").alias("total_sales"))

city_stats.show()
```

```python
city_stats.orderBy(col("total_sales").desc()).show()
```

```python
payment_sales = df.groupBy("payment_method").agg(sum("total_amount").alias("payment_sales"))

payment_sales.show()
```

```python
from pyspark.sql.functions import round

total_sales = df.agg(sum("total_amount")).collect()[0][0]

payment_sales = payment_sales.withColumn(
    "sales_percentage",
    round((col("payment_sales") / total_sales) * 100, 2)
)

payment_sales.show()
```

```python
from pyspark.sql.functions import concat_ws

df = df.withColumn(
    "product_info",
    concat_ws(" - ", col("category"), col("product"))
)
```

```python
daily_sales = df.groupBy("order_date").agg(sum("total_amount").alias("daily_sales")).orderBy("order_date")

daily_sales.show()
```

```python
df.select("order_id","customer_id","total_amount").orderBy(col("total_amount").desc()).show(5)
```
