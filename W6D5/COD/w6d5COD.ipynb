```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as f
```

```python
spark=SparkSession.builder.getOrCreate()
```

```python
e_df=spark.read.csv("employees.csv",header="True",inferSchema="True")
p_df=spark.read.csv("performance.csv",header="True",inferSchema="True")
```

```python
merge_df=e_df.join(p_df,on="employee_id",how="left")
merge_df.show()
```

```python
merge_df.withColumn("full_name",f.coalesce(f.col("full_name"),f.concat_ws(" ","first_name","last_name")))
merge_df.show()
```

```python
df5=merge_df.filter(f.col("rating")<4.0)
df5.select("full_name","department","rating").show()
```
