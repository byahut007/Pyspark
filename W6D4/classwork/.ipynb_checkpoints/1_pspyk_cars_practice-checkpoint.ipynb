```python
#Data
data = [
    ("Toyota", "Camry", 2020, 203, 24425, "Petrol", "Automatic"),
    ("Honda", "Accord", 2019, 192, 23720, "Petrol", "Automatic"),
    ("Ford", "Mustang", 2021, 450, 55000, "Petrol", "Manual"),
    ("Chevrolet", "Malibu", 2018, 160, 22000, "Petrol", "Automatic"),
    ("BMW", "3 Series", 2022, 255, 41250, "Petrol", "Automatic"),
    ("Audi", "A4", 2020, 248, 39900, "Petrol", "Automatic"),
    ("Mercedes", "C-Class", 2021, 255, 43000, "Diesel", "Automatic"),
    ("Hyundai", "Elantra", 2022, 147, 20000, "Petrol", "Manual"),
    ("Kia", "Optima", 2019, 185, 23000, "Petrol", "Automatic"),
    ("Nissan", "Altima", 2020, 188, 24000, "Petrol", "Automatic"),
    ("Tesla", "Model 3", 2022, 283, 46990, "Electric", "Automatic"),
    ("Volkswagen", "Passat", 2018, 174, 22500, "Diesel", "Manual"),
    ("Subaru", "Impreza", 2021, 152, 21000, "Petrol", "Manual"),
    ("Mazda", "Mazda6", 2020, 187, 24500, "Petrol", "Manual"),
    ("Lexus", "ES 350", 2021, 302, 41000, "Hybrid", "Automatic"),
    ("Jaguar", "XE", 2019, 247, 39900, "Diesel", "Automatic"),
    ("Dodge", "Charger", 2020, 292, 31000, "Petrol", "Manual"),
    ("Infiniti", "Q50", 2021, 300, 42000, "Petrol", "Automatic"),
    ("Genesis", "G70", 2022, 252, 39000, "Petrol", "Automatic"),
    ("Acura", "TLX", 2020, 272, 37500, "Petrol", "Automatic")
]
#Reference columns
columns = ["Brand", "Model", "Year", "Horsepower", "Price", "FuelType", "Transmission"]
```

```python
#Create StructType for column input
```

```python
#Import libraries & necessary functions
from pyspark.sql import SparkSession
```

```python
#Create SparkSession
spark = SparkSession.builder.appName("Car Analysis").getOrCreate()
```

```python
#Get first three entried from dataframe
df=spark.createDataFrame(data,columns)
df.show(3)
```

```python
#Print schema and check data types
df.printSchema()
```

```python
#Count total number of cars
df.count()
```

```python
#List all distinct fuel types
df.select("FuelType").distinct().show()
```

```python
#Filter for cars that are Diesel and have Manual transmission
from pyspark.sql import functions as F

df.filter((F.col("FuelType")=="Diesel") & (F.col("Transmission")=="Manual")).show()
```

```python
#Find all Electric cars
df.filter(F.col("FuelType")=="Electric").show()
```

```python
#Count number of cars for each Transmission type
df.groupby("Transmission").count().show()
```

```python
#Calculate average price for each FuelType
df.groupby("FuelType").avg("Price").show()
```

```python
#Show top 5 most expensive cars
df.orderBy(F.desc("Price")).show(5)
```

```python
#Show top 5 most expensive automatic cars
df.filter(F.col("Transmission")=="Automatic").orderBy(F.desc("Price")).show(5)
```

```python
#Create a new column IsPremium if price > 40,000
df=df.withColumn("IsPremium",F.col("Price")>40000)
df.show()
```

```python
#Count number of Hybrid cars
df.filter(F.col("FuelType")=="Hybrid").count()
```

```python
#Find the average horsepower of cars
df.agg(F.avg("Horsepower").alias("Avg_Horsepower")).show()
```

```python
#Find the average horsepower of cars with manual transmission
df1=df.filter(df["Transmission"] == "Manual")
df1.agg(F.avg("Horsepower").alias("Avg_HP_Manual")).show()
```

```python
#Find the average horsepower of cars with manual transmission
df1=df.filter(F.col("Transmission") == "Manual")
df1.agg(F.avg("Horsepower").alias("Avg_HP_Manual")).show()
```

```python
#Find cars with price between 30,000 and 45,000
df.filter(F.col("Price").between(30000, 45000)).show()
```

```python
#Find number of cars released each year
df.groupBy("Year").count().orderBy("Year").show()
```

```python
#Add column Age of car assuming current year is 2025
df = df.withColumn("Age", F.lit(2025) - F.col("Year"))
df.show()

df.withColumn("Age",F.lit(2025)-F.col("Year"))
```
